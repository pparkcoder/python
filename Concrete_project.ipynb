# 모든 coulmns의 Outlier 제거
### IQR = q3(75%) - q1(25%)
### Outlier의 범위  : q1 - (1.5 * IQR) > X | X > q3 + (1.5 * IQR)

```python
import numpy as np
def get_outlier_index(df, columns) : # outlier에 해당하는 index 반환 함수
    temp = df[columns]
    q1 = np.percentile(temp, 25)
    q3 = np.percentile(temp, 75)
    iqr = q3 - q1
    lowest = q1 - (1.5 * iqr)
    highest = q3 + (1.5 * iqr)
    return temp[(lowest > temp) | (temp > highest)].index 

print(concrete.shape)
for i in concrete.columns : 
        if (i == 'strength') : continue
        out_lier_index = get_outlier_index(concrete,i)
        concrete.drop(out_lier_index,axis = 0, inplace = True)
print(concrete.shape) # Outlier 제거 후 concrete shape
```

### LinearRegression

### RandomForest - 회귀, 분류 둘 다 가능

```python
from sklearn.model_selection import train_test_split
data = concrete.drop(['strength'], axis = 1)
label = concrete['strength']
x_train, x_test, y_train, y_test = train_test_split(data, label, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

pipe = Pipeline([('preprocessing', StandardScaler()), ('classifier', Ridge())])
param_grid = [
    {'classifier': [LinearRegression()], 'preprocessing': [StandardScaler(),MinMaxScaler(),None],
             'classifier__fit_intercept':[True,False],
             'classifier__n_jobs' :[100,1,10,2,5,200,300,500,1000],
             'classifier__normalize':[True,False]},
    {'classifier': [KNeighborsRegressor()], 'preprocessing': [StandardScaler(),MinMaxScaler(),None],
             'classifier__n_neighbors':[3,5,7,10],
             'classifier__n_jobs' :[1,2,3],
             'classifier__weights':['uniform','distance']},
    {'classifier': [RandomForestClassifier()], 'preprocessing': [None],
             'classifier__min_samples_leaf':[0.5,1,2],
             'classifier__n_estimators': [20,100,200],
             'classifier__criterion':['gini','entropy'],
             'classifier__bootstrap' :[True,False],
             'classifier__oob_score':[True,False]},
    {'classifier': [Lasso()],
     'preprocessing': [MinMaxScaler(),StandardScaler(),None],'classifier__fit_intercept':[True,False],
             'classifier__alpha' :[1.0,0.0001,0.1,0.001],
             'classifier__max_iter' :[500,2000,5000,20000,50000],
             'classifier__normalize':[True,False]},
    {'classifier': [Ridge()],
     'preprocessing': [MinMaxScaler(),StandardScaler(),None],'classifier__fit_intercept':[True,False],
             'classifier__alpha' :[1.0,0.0001,0.1,0.001],
             'classifier__max_iter' :[500,2000,5000,20000,50000],
             'classifier__normalize':[True,False]}]

grid = GridSearchCV(pipe, param_grid, cv=5)
grid.fit(x_train, y_train)

print("최적의 매개변수:\n{}\n".format(grid.best_params_))
print("최상의 교차 검증 점수: {:.2f}".format(grid.best_score_))
print("테스트 세트 점수: {:.2f}".format(grid.score(x_test, y_test)))
```
